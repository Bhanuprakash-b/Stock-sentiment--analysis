```python
import pandas as pd
```

```python
df=pd.read_csv('Combined_News_DJIA.csv', encoding = "ISO-8859-1")
```

```python
df2=df[480:]
```

```python
train=df2[df2['Date']<'20150101']
test=df2[df2['Date']>'20140101']
```

```python
#removing punctuations
data=train.iloc[:,2:27]
```

```python
data
```

```python
data.replace("[^a-zA-Z]"," ")
```

```python
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
str(1)
```

```python
#lowering
for i in new_Index:
    data[i]=data[i].str.lower()
data.head()
```

```python
#combining all the sentences 
```

```python
''.join(str(x) for x in data.iloc[1,0:25])
```

```python
#iterating and combing all the records
headlines=[]
for i in range(0,len(data.index)):
    headlines.append(''.join(str(x) for x in data.iloc[i,0:25]))
```

```python
headlines[1]
```

```python
df2.head(2)
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2))
#example the cat is stting.
#coverting the above statement into vectors on keeping whole dataset in view
#ngram_range is used to club two words as per our instruction
#the cat->true
#the dog->true
#ello->false

```

```python
traindata=cv.fit_transform(headlines)
```

```python
print(traindata)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(n_estimators=200,criterion='entropy')
rc.fit(traindata,train['Label'])
```

```python
testdata=[]
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = cv.transform(test_transform)
predictions = rc.predict(test_dataset)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_m=confusion_matrix(test['Label'],predictions)
print(confusion_m)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(test['Label'],predictions)
print(accuracy)
c=classification_report(test['Label'],predictions)
```

```python
rc2=RandomForestClassifier(n_estimators=200,criterion='entropy')
rc2.fit(tr2,train['Label'])
```

```python
td2 = tv.transform(test_transform)
p2=rc2.predict(td2)
```
